```python
import pandas as pd
import tkinter as tk
import warnings 

warnings.filterwarnings('ignore')
```

```python
app = tk.Tk()
app.title('House Price Predictor')
app.geometry('600x300')
app.configure(bg = 'grey')


df = pd.read_csv('housing.csv')
x = df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Area Population']]
y = df['Price']
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state=10)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
model.coef_
pd.DataFrame(model.coef_,columns=['coef.'], index=x.columns)
ypred = model.predict(xtest)
from sklearn.metrics import mean_absolute_error
final_model = model
model = pd.to_pickle(final_model,'HousePricePredictor')
model = pd.read_pickle('HousePricePredictor')




tk.Label(app, text = 'Income:',font ='b').place(x=25,y=25)
tk.Label(app, text = 'House Age:',font ='b').place(x=25,y=65)
tk.Label(app, text = 'Number of Rooms:',font ='b').place(x=25,y=110)
tk.Label(app, text = 'Population:',font ='b').place(x=25,y=155)


income_var = tk.StringVar(app)
house_var = tk.StringVar(app)
rooms_var = tk.StringVar(app)
population_var = tk.StringVar(app)


tk.Entry(app, textvariable = income_var, font ='b').place(x=170,y=25)
tk.Entry(app, textvariable = house_var, font ='b').place(x=170,y=65)
tk.Entry(app, textvariable = rooms_var, font ='b').place(x=170,y=110)
tk.Entry(app, textvariable = population_var, font ='b').place(x=170,y=155)


def predict():
    query = {'Avg. Area Income':[],'Avg. Area House Age':[],'Avg. Area Number of rooms':[],'Area Population':[]}
    df = pd.DataFrame(query)
    f=open('query.csv','a')
    data=[income_var.get(),house_var.get(),rooms_var.get(),population_var.get()]
    f.write(','.join(data)+'\n')
    f.close()

        
    cols = ['Avg. Area Income','Avg. Area House Age','Avg. Area Number of rooms','Area Population']
    
    
    
    income = income_var.get()
    house_age = house_var.get()
    rooms = rooms_var.get()
    population = population_var.get()


    
    query = pd.DataFrame({'Avg. Area Income':[income],
                     'Avg. Area House Age':[house_age],
                     'Avg. Area Number of Rooms':[rooms],
                     'Area Population':[population]})
    
    
    predict_var = tk.Variable(app)
    predict_var.set('%.1f$'%model.predict(query)[0])
    tk.Label(app, text = 'House Price =', font = 'b' ,bg = 'blue',fg = 'white').place(x=25,y=250)
    predict = tk.Label(app, textvariable = predict_var, font = 'b',bg='white',width = 20).place(x=170,y=250)
    


    
 


tk.Button(app, text = 'Prection Price', font = 'b' ,bg = 'red',fg = 'white',command = predict ).place(x=200,y=200)


app.mainloop()
```
