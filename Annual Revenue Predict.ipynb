```python
x = df[['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10','P11','P12','P13','P14','P15','P16','P17','P18','P19','P20','P21','P22','P23','P24','P25','P27','P28','P29','P30','P31','P32','P33','P34','P35','P36','P37']]
  
y = df['revenue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.25, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_,columns=['coef.'],index = x.columns)
```

```python
print('%.1f%%'%(model.score(xtest,ytest)*100))
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
100 * mean_absolute_error(ytest,ypred)/ypred[0]
```

```python
final_model = model
model = pd.to_pickle(final_model,'Annual Price Predictor')
model
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
x_poly = PolynomialFeatures().fit_transform(X)
```

```python
x_poly2 = PolynomialFeatures(degree=2).fit_transform(X)
```

```python
model2.score(xtest2, ytest2)
```
